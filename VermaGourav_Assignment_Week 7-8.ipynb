### Finding Missing Value

```python
# Finding Missing value counts for each column

met.isna().sum().sort_values(ascending = False)
```

```python
# Find missing value percentage
met.isna().mean().sort_values(ascending = False).round(4)*100
```

### Remove Duplicates

```python
met.duplicated(['Object Number']).sum()
```

It seems we have 2576 duplicate entries in the file. Lets drop them.

```python
# Remove Duplicates
metd = met.drop_duplicates(subset='Object Number')
metd.shape
```

### Create Hierarchical Index

```python
# Create a small dataframe containing selected columns

metn = metd[['Object Number', 'Department', 'Title', 'Object Name', 'Medium', 'Classification', 'Is Public Domain']]

# Select random sample of 1000 rows
metr = metn.sample(1000)

# Reset Index
metr = metr.reset_index()
```

```python
# Create Hierarchical indexing, using columns department and medium

meth = metr.set_index(['Department', 'Medium'])
meth.head()
```

### Reshaping

```python
# Reshaping the sample dataframe. Changing columns into rows.
st = metr.stack()
st
```

```python
# Reshape from the rows into the columns
st.unstack()
```

### Grouping with Functions

```python
# Lets find whats the count if the object is public domain or not.

metg = metr.set_index(['Is Public Domain'])
metg.groupby(bool).count()
```

### Crosstab

```python
# Summarize the total records by calssifications and departments, and display top 10 classifications

met_ic = pd.crosstab(metn.Classification, metn.Department, margins = True)
met_ic.sort_values('All', ascending=False).head(11)
```
